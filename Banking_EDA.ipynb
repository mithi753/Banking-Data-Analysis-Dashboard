```python
import pandas as pd
data=pd.read_csv('Banking.csv')
```

```python
data.describe()
```

```python
data['Estimated Income'].value_counts()
```

```python
bins=[0,100000,300000,float('inf')]
labels=['Low','Medium','High']
data['Estimated Income']=pd.cut(data['Estimated Income'],bins=bins,labels=labels)
```

```python
data['Estimated Income'].value_counts().plot(kind='bar')
```

```python
#examine the distribution of unique categories
categorical_col=['Nationality','Fee Structure','Loyalty Classification','Amount of Credit Cards','Properties Owned','Risk Weighting','BRId', 'GenderId','IAId','Estimated Income']
```

```python
for col in categorical_col:
  print(f"{col}")
  display (data[col].value_counts())
```

###Univariate Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt
for i,predictor in enumerate(categorical_col):
  plt.figure(i)
  sns.countplot(data=data,x=predictor,hue='GenderId')
```

### Bivariate Analysis

```python
for i,predictor in enumerate(categorical_col):
  plt.figure(i)
  sns.countplot(data=data,x=predictor,hue='Nationality')
```

```python
#Histplot of valuse counts for different occupation
for col in categorical_col:
  if col == 'Occupation':
    continue
  plt.figure(figsize=(8,4))
  sns.histplot(data[col])
  plt.title(f"Distribution of {col}")
  plt.xlabel(col)
  plt.ylabel("Frequency")
  plt.show()
```

###Numerical Analysis

```python
 numerical_col=['Estimated Income','Superannuation Savings']
 #Univariate analysis and visualization
 plt.figure(figsize=(8,4))
 for i,col in enumerate(numerical_col):
   plt.subplot(2,1,i+1)
   sns.histplot(data[col],kde=True)
   plt.title(f"Distribution of {col}")
   plt.ylabel("Frequency")
plt.show()
```

###Heatmap

```python
numerical_col=['Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']
correlation_matrix=data[numerical_col].corr()
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm',fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

## Insights
1. Here  we can see a strong correlation among 'Bank deposits' with "Checking accounts' , 'savings accounts' and 'foreign currency account'  indicating that customers who maintain high balance in  one account type often  held substantial amount across other  account as well .
